```python
def get_api_key(configFile = 'configs/googleMaps_API_key.json'):
    with open(configFile) as config:
        config_dict = json.load(config) 
    return config_dict['apiKey_rck']
```

```python
def code_city(city):
    googleMapsKey = get_api_key()
    client = googlemaps.Client(key=googleMapsKey)
    request = client.geocode(city)
    return [request[0]['geometry']['location']['lat'], request[0]['geometry']['location']['lng']]
```

```python
df = pd.read_excel('data/nfl_lookup_table.xlsx')
df['LatLong'] = df['City'].map(code_city) 
df[['Latitude', 'Longitude']] = pd.DataFrame(df['LatLong'].tolist(), index = df.index)
df.drop('LatLong', axis=1, inplace=True)
display(df)
df.to_csv('data/nfl_lookup_table.csv', index=False)
```
